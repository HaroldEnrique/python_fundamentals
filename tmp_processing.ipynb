```python
import snowballstemmer
import spacy

def apply_lemmatization(text):
    nlp = spacy.load("es_core_news_sm")
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])

def apply_stemming(text):
    stemmer = snowballstemmer.stemmer('spanish');
    words = stemmer.stemWords(text.split())
    phrase = ' '.join(words)
    return phrase
```

```python
print(apply_lemmatization("corriendo a la playa va un perrito, cuyo dueño se durmió por tanto cansancio"))
print(apply_lemmatization("ayer te vi comiendo un rico mango en la esquina del parque"))
print(apply_lemmatization("día como hoy salimos a la playa a surfear, a disfrutar de las olas de la vida"))
```

```python
print(apply_stemming("corriendo a la playa va un perrito, cuyo dueño se durmió por tanto cansancio"))
print(apply_stemming("ayer te vi comiendo un rico mango en la esquina del parque"))
print(apply_stemming("día como hoy salimos a la playa a surfear, a disfrutar de las olas de la vida"))
```

```python
print(apply_lemmatization(apply_stemming("corriendo a la playa va un perrito, cuyo dueño se durmió por tanto cansancio.")))
print(apply_lemmatization(apply_stemming("ayer te vi comiendo un rico mango en la esquina del parque")))
print(apply_lemmatization(apply_stemming("un día como hoy salimos a la playa a surfear, a disfrutar de las olas de la vida")))
```

```python
from text_processing.text_preprocessor import TextPreprocessor
import pandas as pd

df=pd.read_csv('H:/courses/dmc_mle/practica_python/tweets.csv')
df['content'] = df['content'].astype(str)
```

```python
preprocessor = TextPreprocessor(remove_links=True, 
                                remove_hastags=True,
                                remove_characters=True, 
                                convert_to_lowercase=True, 
                                remove_emojis=True,
                                remove_numbers=True,
                                remove_stopwords_flag= True,
                                apply_lemmatization=True,
                                apply_stemming=True)
df_sample = df.iloc[1:50,:].copy()
df_sample['content_2'] = df_sample['content'].apply(preprocessor.preprocess_text)
```
